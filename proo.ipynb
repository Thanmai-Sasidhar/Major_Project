```python
# Cell 1: Imports
import pandas as pd
import numpy as np
from datetime import datetime, timedelta
import torch
import torch.nn.functional as F
from torch_geometric.nn import GATConv, global_mean_pool
from torch_geometric.data import Data
import plotly.graph_objects as go
from skyfield.api import load, EarthSatellite
from sgp4.api import Satrec, WGS72
import matplotlib.pyplot as plt
import networkx as nx

print("PyTorch version:", torch.__version__)
device = torch.device('cpu')
print("Using device:", device)
```

```python
# ================================
# Stage 1: Load + SGP4 Propagation + Visualization
# ================================

import pandas as pd
import numpy as np
from datetime import datetime
from skyfield.api import load, EarthSatellite, utc
from sgp4.api import Satrec, WGS72
import plotly.graph_objects as go

# -------------------------------
# Load dataset
# -------------------------------
df = pd.read_csv("data/space_objects.csv")
print(f"Total objects in dataset: {len(df)}")

# Use small sample for testing
sample_df = df.head(25).copy()

# -------------------------------
# Time setup (FIXED)
# -------------------------------
ts = load.timescale()

minutes = list(range(0, 1440, 10))  # every 10 mins
times = ts.utc(2026, 3, 1, 0, minutes)

satellites = []
positions = {}

# -------------------------------
# Load satellites + propagate
# -------------------------------
for _, row in sample_df.iterrows():
    try:
        # Skip bad rows
        if pd.isna(row['EPOCH']):
            continue

        satrec = Satrec()

        # FIX 1: Proper timezone handling
        epoch_dt = datetime.fromisoformat(row['EPOCH'].replace('Z', '+00:00'))
        epoch_dt = epoch_dt.replace(tzinfo=utc)
        jd = ts.from_datetime(epoch_dt).tt

        # FIX 2: sgp4init argument corrected ('U' instead of 0)
        satrec.sgp4init(
            WGS72,
            'U',   # classification FIXED
            int(row['NORAD_CAT_ID']),
            jd,
            float(row['BSTAR']),
            float(row['MEAN_MOTION_DOT']),
            0.0,
            float(row['ECCENTRICITY']),
            np.deg2rad(float(row['ARG_OF_PERICENTER'])),
            np.deg2rad(float(row['INCLINATION'])),
            np.deg2rad(float(row['MEAN_ANOMALY'])),
            float(row['MEAN_MOTION']) * 2 * np.pi / 1440.0,
            np.deg2rad(float(row['RA_OF_ASC_NODE']))
        )

        sat = EarthSatellite.from_satrec(satrec, ts)
        sat.name = row['OBJECT_NAME']
        satellites.append(sat)

        # Propagate orbit
        pos_list = []
        for t in times:
            geocentric = sat.at(t)
            pos_list.append(geocentric.position.km)

        positions[sat.name] = np.array(pos_list)

        print(f"✓ Loaded: {sat.name}")

    except Exception as e:
        print(f"✗ Failed {row.get('OBJECT_NAME', 'Unknown')}: {e}")

print(f"\n✅ Successfully loaded {len(satellites)} satellites with SGP4")

# -------------------------------
# 3D Visualization
# -------------------------------
fig = go.Figure()

# Earth sphere
u, v = np.mgrid[0:2*np.pi:60j, 0:np.pi:30j]
fig.add_trace(go.Surface(
    x=6371*np.cos(u)*np.sin(v),
    y=6371*np.sin(u)*np.sin(v),
    z=6371*np.cos(v),
    colorscale='Blues',
    opacity=0.2,
    showscale=False,
    name='Earth'
))

# Plot satellite orbits
colors = ['red', 'blue', 'green', 'orange', 'purple', 'cyan']

for i, sat in enumerate(satellites):
    pos = positions[sat.name]

    fig.add_trace(go.Scatter3d(
        x=pos[:, 0],
        y=pos[:, 1],
        z=pos[:, 2],
        mode='lines',
        name=sat.name,
        line=dict(color=colors[i % len(colors)], width=3)
    ))

# Layout
fig.update_layout(
    title="SGP4 Propagated Orbits",
    scene=dict(aspectmode='cube'),
    width=950,
    height=700
)

fig.show()
```

```python
import torch
from torch_geometric.data import Data

# Node features (positions)
x = torch.tensor(final_positions, dtype=torch.float)

# Edges
edge_index = torch.tensor(edges, dtype=torch.long).t().contiguous()

# Labels (node-level)
labels = []

for i in range(len(sat_names)):
    risk = 0
    for j in range(len(sat_names)):
        if i != j and distance_matrix[i][j] < 1000:
            risk = 1
    labels.append(risk)

y = torch.tensor(labels, dtype=torch.float)

data = Data(x=x, edge_index=edge_index, y=y)

print(data)
```

```python
from torch_geometric.nn import GATConv
import torch.nn.functional as F

class GATModel(torch.nn.Module):
    def __init__(self):
        super().__init__()
        self.gat1 = GATConv(3, 16)
        self.gat2 = GATConv(16, 1)

    def forward(self, data):
        x, edge_index = data.x, data.edge_index
        
        x = self.gat1(x, edge_index)
        x = F.relu(x)
        
        x = self.gat2(x, edge_index)
        
        return x.squeeze()

model = GATModel()
```

```python
optimizer = torch.optim.Adam(model.parameters(), lr=0.01)

for epoch in range(100):
    optimizer.zero_grad()
    
    out = model(data)
    
    loss = F.mse_loss(out, data.y)
    loss.backward()
    optimizer.step()
    
    if epoch % 20 == 0:
        print(f"Epoch {epoch}, Loss: {loss.item():.4f}")
# =========================
# SAVE MODEL
# =========================
torch.save(model.state_dict(), "gat_model.pth")

print("✅ Model saved as gat_model.pth")
```
